# Sentiment Analysis for Stock

```python
import pandas as pd
```

```python
csv_file = pd.read_csv(r"C:\Users\Admin\Automation_Python\NLP Projects\Data.csv",encoding= 'ANSI')
```

```python
#Splitting large chunk of data to Training dataset. Smaller one for Testing
train = csv_file[csv_file['Date'] < '2015-01-01']
test = csv_file[csv_file['Date'] > '2014-12-31']
```

```python
train
```

```python
#Feature engineering
#Removing special characters and changing everything to lowercase

data = train.iloc[:,2:27]
data.replace("[^A-Za-z]",' ',regex = True,inplace = True)
```

```python
#Renaming column names as index for readable format
lst = [i for i in range(25)]
lst

#Assigning generated values as index name list
col_name = [str(x) for x in lst]
col_name

data.columns = col_name

data
```

```python
#Converting all data into lower
for z in col_name:
    data[z] = data[z].str.lower()
data
```

```python


" ".join(str(x) for x in data.iloc[0,0:25])
```

```python
#Appending all columns to a single sentence for all rows
sentences = []
for row in range(len(data.index)):
    sentences.append(" ".join(str(x) for x in data.iloc[row,0:25]))
```

```python
sentences[3000]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
#Implementing Bag of Words
cv = CountVectorizer(ngram_range=(5,5))
traindataset = cv.fit_transform(sentences)
```

```python
#Implementing RandomForest Classifier
randomclassifier = RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
```

```python
test_transform= []
for row in range(len(test.index)):
    test_transform.append(" ".join(str(x) for x in test.iloc[row,0:27]))
test_dataset = cv.transform(test_transform)
predictions = randomclassifier.predict(test_dataset)
```

```python
#Checking Accuracy for the predictions
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
matrix = confusion_matrix(test['Label'],predictions)
print(matrix)

score = accuracy_score(test['Label'],predictions)
print(score)

report = classification_report(test['Label'],predictions)
print(report)
```
